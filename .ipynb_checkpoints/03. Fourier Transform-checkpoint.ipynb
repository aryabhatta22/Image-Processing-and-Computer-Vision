### Fourier Transform: 

The Fourier transform allows us to represent an image by its frequency spectrum, and by doing so, we can remove noise which have a repeating pattern to it.

# 1. Importing libraries and images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/profile_color.jpeg')
image_copy = np.copy(image)
plt.imshow(image_copy)
```

```python
#image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(image_copy, cmap = 'gray')
```

# 2. Fourier Transform of Image

```python
f = np.fft.fft2(image_copy)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.imshow(magnitude_spectrum, cmap = 'gray')
```

# 3. Plotting results

```python
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.set_title('Original Image')
ax1.imshow(image_copy,cmap ='gray')

ax2.set_title('Fourier Transform')
ax2.imshow(magnitude_spectrum,cmap ='gray')

fig.tight_layout()
fig.savefig('results/fourier_transform.png')
```
