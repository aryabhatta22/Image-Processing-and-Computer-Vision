# 1. Importing libraries and images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/profile.jpg')
plt.imshow(image, cmap= 'gray')
```

# 2. Applying Gaussian blur

```python
image_copy = np.copy(image)
gray_blur = cv2.GaussianBlur(image_copy, (5,5), 5)
plt.imshow(gray_blur, cmap ='gray')
```

# 3. Plotting images

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (8,5))

ax1.set_title('Orignal Image')
ax1.imshow(image_copy, cmap = 'gray')

ax2.set_title('Gaussian Blured Image')
ax2.imshow(gray_blur, cmap ='gray')

fig.tight_layout()

fig.savefig('results/low_pass_filter.png')
```
