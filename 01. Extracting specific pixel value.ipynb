# 1. Importing libraries and image

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = mpimg.imread('images/profile_color.jpeg')
plt.imshow(image)
```

# 2. Extracting information & converting to grayscale

```python
print("Image Dimension: ", image.shape)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap = "gray")
```

# 3. Extracting pixel value

```python
x = 190
y = 375

print("pixel value of gray image at  x = 190 and y = 375 :")
print(gray_image[y,x])
```
