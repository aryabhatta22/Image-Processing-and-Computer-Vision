# 1. Importing Libraries and image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/profile_color.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
```

# 2. Applying High Pass Filters

```python
# For horizontal edges
hor = np.array([[-1,-2,-1],
                    [0,0,0],
                    [1,2,1]])

# for vertical edges
ver = np.array([[-1,0,1],
                    [-2,0,2],
                    [-3,0,3]])
```

```python
hor_edge = cv2.filter2D(gray_img, -1, hor)
plt.imshow(hor_edge, cmap='gray')
```

```python
ver_edge = cv2.filter2D(gray_img, -1, ver)
plt.imshow(ver_edge, cmap='gray')
```

# 3. Plotting Images

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (8,8))

ax1.set_title('Original Image')
ax1.imshow(image)

ax2.set_title('Horizontal Edges')
ax2.imshow(hor_edge, cmap ='gray')

ax3.set_title('Vertical Edges')
ax3.imshow(ver_edge, cmap ='gray')

fig.tight_layout()
fig.savefig('results/high_pass_filter.png')
```
