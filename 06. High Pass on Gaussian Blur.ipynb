# 1. Importing Libraries and image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/profile_color.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# colored image to gray image
image_copy = np.copy(image)
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

# 2. Applying filters

```python
# vertical filter Without blur
ver_edge = np.array([[-2,0,2],
                    [-1,0,1],
                    [-2,0,2]])
filtered_image = cv2.filter2D(gray, -1, ver_edge)
plt.imshow(filtered_image, cmap='gray')
```

```python
# Gaussian Blur on image

filtered_gausssian = cv2.GaussianBlur(gray, (5,5), 0)
plt.imshow(filtered_gausssian, cmap='gray')
```

```python
# Vertical edges on blured image

Blurred_edges = cv2.filter2D(filtered_gausssian, -1, ver_edge)
plt.imshow(Blurred_edges, cmap='gray')
```

# 3. Plotting Images

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize =(10,5))

ax1.set_title('Original Image')
ax1.imshow(image)

ax2.set_title('Vertical Edge without Blur')
ax2.imshow(filtered_image, cmap='gray')

ax3.set_title('Vertical Edge after Blur')
ax3.imshow(Blurred_edges, cmap='gray')

fig.tight_layout()

fig.savefig('results/high_pass_on_blur.png')
```
