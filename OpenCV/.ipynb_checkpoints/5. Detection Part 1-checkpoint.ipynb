```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread('../images/group.jpeg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

# Face Detection

```python
face_classifier = cv2.CascadeClassifier('HaarCascade/haarcascade_frontalface_default.xml')
```

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

```python
if faces is ():
    print('No face Found')
```

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img, (x,y), (x+w,y+h), (255,0,0), 10)
    return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```

# Eye Detection

```python
eye_classifier = cv2.CascadeClassifier('HaarCascade/haarcascade_eye.xml')
```

```python
def detect_eyes(fix_img):
    eyes_rects = eye_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(fix_img, (x,y), (x+w, y+h), (255,255,255), 10)
    return fix_img
```

```python
result =detect_eyes(fix_img)
plt.imshow(result)
```

# Faces and Eyes

```python
def detect_face_and_eyes(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img, 1.3, 5)     # tune last two parameters
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img, (x,y), (x+w,y+h), (255,0,0), 7)
        eyes_rects = eye_classifier.detectMultiScale(fix_img)
        for (ix,iy,iw,ih) in eyes_rects:
            cv2.rectangle(fix_img, (ix,iy), (ix+iw, iy+ih), (255,255,255), 5)
    return fix_img
```

```python
result = detect_face_and_eyes(fix_img)
plt.imshow(fix_img)
```
