### Importing libraries and video

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
body_classifier = cv2.CascadeClassifier('HaarCascade/haarcascade_fullbody.xml')
```

```python
cap = cv2.VideoCapture('../Videos/pedestrian.mp4')
```

### Function to rescale the video frame

```python
def rescale_frame(frame, percent=75):
    width = int(frame.shape[1] * percent/ 100)
    height = int(frame.shape[0] * percent/ 100)
    dim = (width, height)
    return cv2.resize(frame, dim, interpolation =cv2.INTER_AREA)
```

### Detecting full body in the video frame

```python
while cap.isOpened():
    #Read the capture
    ret, frame = cap.read()
    frame = rescale_frame(frame, percent=25)  # change percentage as per video resolution
    #Pass the frame to the classifier
    bodies = body_classifier.detectMultiScale(frame, 1.2, 3)
    
    if ret == True:
        #Bounding Boxes for identified bodies
        for (x,y,w,h) in bodies:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (25,255,125), 5)
            cv2.imshow('Pedestrians', frame)
        # Exit with ESC key
        if cv2.waitKey(1) == 27:
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
    
```

# Detecting moving cars

```python
car_classifier = cv2.CascadeClassifier('HaarCascade/haarcascade_cars.xml')
```

```python
cap = cv2.VideoCapture('../Videos/pedestrian.mp4')
```

```python
while cap.isOpened():
    #Read the capture
    ret, frame = cap.read()
    frame = rescale_frame(frame, percent=25)  # change percentage as per video resolution
    #Pass the frame to the classifier
    cars = car_classifier.detectMultiScale(frame, 1.4, 2)
    
    if ret == True:
        #Bounding Boxes for identified bodies
        for (x,y,w,h) in cars:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (25,255,125), 5)
            cv2.imshow('Cars', frame)
        # Exit with ESC key
        if cv2.waitKey(1) == 27:
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
    
```

# Car's Plate Detection

```python
img = cv2.imread('../images/car_plate.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
plate_classifier = cv2.CascadeClassifier('./HaarCascade/haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    plate_img = img.copy()
    plate_rects = plate_classifier.detectMultiScale(plate_img, 1.1, 1)
    for(x,y,w,h) in plate_rects:
        cv2.rectangle(plate_img, (x,y), (x+w,y+h),(255,255,255),5)
    return plate_img
```

```python
result = detect_plate(img)
display(result)
```
