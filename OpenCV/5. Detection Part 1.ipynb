```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread('../images/group.jpeg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

# Face Detection

```python
face_classifier = cv2.CascadeClassifier('HaarCascade/haarcascade_frontalface_default.xml')
```

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

```python
if faces is ():
    print('No face Found')
```

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img, (x,y), (x+w,y+h), (255,0,0), 10)
    return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```
